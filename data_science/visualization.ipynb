visualization
    - single column analysis -> univariate analysis
    - two columns analysis -> bivariate analysis
    - multiple columns analysis -> multivariate analysis
- types of visualization
    - line plot (time series, trend, seasonality, cyclicity, noise, outliers, missing values, etc.)
    - area plot (similar to line plot, but area under the line is emphasized)
    - bar plot (categorical data, nominal, ordinal, etc. for discrete data comparison)
    - histogram (continuous data, frequency, distribution, etc. for continuous data comparison)
    - pie chart (same as bar chart but smaller number of categories and percentage comparison)
    - box plot  (statistical evaluation,outliers, quartiles, etc. for continuous data comparison)
    - scatter plot (correlation, relationship, etc. for continuous data comparison)
    - bubble plot (scatter plot with additional dimension of data)

```python
data['population'].plot(figsize=(12,4), style='sr--', 
                        markersize=15, linewidth=2, 
                        title='Poupulation of states', 
                        xlabel='States', 
                        ylabel='Population')
plt.xticks(data.index, data['state'], rotation=45) # this makes the xticks to be the state names
plt.show() # this just shows the plot
```

```python
data['rainfall'].plot(kind='area', figsize=(12,4), color='red', alpha=.5, title='Rainfall in states')
data['rainfall'].plot(style='ok-', markerfacecolor='black')
plt.xticks(data.index, data['state'])
plt.show()
```

```python
data['rainfall'].plot(kind='area', figsize=(12,4), color='red', alpha=.1, title='Rainfall in states')
data['rainfall'].plot(kind='bar', color='green', alpha=.5, width=.2)
```

```python
# bar with text
data['rainfall'].plot(kind='bar', color='green', alpha=.5)
plt.ylim(0,100) # explicitly set the y axis range to 0 to 100
for i in range(len(data['rainfall'])):
    plt.text(i, data['rainfall'][i], data['rainfall'][i], ha='center', va='bottom')
```

```python
# bar with text using bar_label (latest)
graph = data['rainfall'].plot(kind='bar', color='green', alpha=.5)
plt.ylim(0,100)
graph.bar_label(graph.containers[0], label_type='edge')
plt.show()
```

```python
ax = data[['temperature','rainfall']].plot(kind='bar', figsize=(12,4), 
                                           width=.8, color=['red','blue'], alpha=.8)
ax.set_xticklabels(data['state'])
ax.bar_label(ax.containers[0], label_type='center', padding=3)
ax.bar_label(ax.containers[1], label_type='center', padding=3, color='white')
plt.show()
```
